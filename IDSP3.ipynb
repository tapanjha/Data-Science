**Prgram-3(a) To perform sorting on series data**

Series is a one-dimensional labeled array capable of holding data of the type integer, string, float, python objects, etc. The axis labels are collectively called index. 

Now, Let’s see a program to sort a Pandas Series.

For sorting a pandas series the Series.sort_values() method is used.

Pandas **Series.sort_values()** function

Pandas Series.sort_values() function is used to sort the given series object in ascending or descending order by some criterion. The function also provides the flexibility of choosing the sorting algorithm.

**Syntax: Series.sort_values(axis=0, ascending=True, inplace=False, kind=’quicksort’, na_position=’last’)**

Parameter :
**axis** : Axis to direct sorting.

**ascending** : If True, sort values in ascending order, otherwise descending.

**inplace** : If True, perform operation in-place.

**kind** : Choice of sorting algorithm.

**na_position** : Argument ‘first’ puts NaNs at the beginning, ‘last’ puts NaNs at the end.

**Returns** : Series

**Example #1:** Use Series.sort_values() function to sort the elements of the given series object in descending order.

```python
# importing pandas as pd
import pandas as pd

# Creating the Series
sr = pd.Series([19.5, 16.8, 22.78, 20.124, 18.1002])

# Print the series
print(sr)
```

```python
# sort the values in Ascending order
sr.sort_values(ascending = True)
```

```python
# sort the values in descending order
sr.sort_values(ascending = False)
```

```python
import pandas as pd
series = pd.Series(['JAVA','PySpark','Spark','Hadoop'])
print(series)
```

```python
# Sort the series in ascending order.
#sortedseries = series.sort_values() # default
sortedseries = series.sort_values(ascending=True)
print(sortedseries)
```

**Prgram-3(b) To perform sorting on dataframe**

```python
# importing pandas library
import pandas as pd

# creating and initializing a nested list
age_list = [['Afghanistan', 1952, 8425333, 'Asia'],
			['Australia', 1957, 9712569, 'Oceania'],
			['Brazil', 1962, 76039390, 'Americas'],
			['China', 1957, 637408000, 'Asia'],
			['France', 1957, 44310863, 'Europe'],
			['India', 1952, 3.72e+08, 'Asia'],
			['United States', 1957, None, 'Americas']]

# creating a pandas dataframe
df = pd.DataFrame(age_list, columns=['Country', 'Year',
									'Population', 'Continent'])

df
```

```python
# Sorting by column 'Country'
df.sort_values(by=['Country'])
```

```python
# Sorting by column "Population"
df.sort_values(by=['Population'], ascending=False)
```

```python
# Sorting by column "Population"
# by putting missing values first
df.sort_values(by=['Population'], na_position='first')
```

```python
# Sorting by columns "Continent"
df.sort_values(by=['Continent'])
```

**Sort rows or columns in Pandas Dataframe based on values**

```python
# import pandas library as pd
import pandas as pd

# List of Tuples
students = [('Ankita', 22, 'Maths', 76 ),
		('Ankita', 31, 'Physics', 65),
		('Rahul', 16, 'Hindi', 56),
		('Simran', 41, 'Hindi', 78),
		('Shaurya', 33, 'English', 75),
		('Harshita', 35, 'Chemistry',82),
		('Swapnil', 35, 'Maths', 92),
		('Priya', 35, 'Physics', 87),
		('Jeet', 35, 'Maths', 59),
		('Ananya', 35, 'Physics', 70)
			]

# Create a DataFrame object from
# list of tuples with columns
# and indices.
df= pd.DataFrame(students, columns =['Name', 'Age',
										'Subject', 'Marks'],
						index =[ 'a', 'b', 'c', 'd', 'e',
								'f', 'g', 'h', 'i', 'j'])
# show the dataframe
df
```

```python
# Sort the rows of dataframe by 'Name' column
rslt_df = df.sort_values(by = 'Name')
  
# show the resultant Dataframe
rslt_df
```

```python
#sort Dataframe rows based on "Name' 
# column in Descending Order
rslt_df = df.sort_values(by = 'Name', ascending = False)
  
# show the resultant Dataframe
rslt_df
```

```python
# sort Dataframe rows based on a 'Name' & 'Age' columns
  
# if duplicate value is present in 'Name' column
# then sorting will be done according to 'Age' column
rslt_df = df.sort_values(by = ['Name', 'Age'])
  
# show the resultant Dataframe
rslt_df
```

**Sort a DataFrame based on row index**

```python
# import pandas library as pd
import pandas as pd

# sort the rows of dataframe
# based on row index
rslt_df = df.sort_index()

# show the resultant Dataframe
rslt_df
```

```python
# import pandas library as pd
import pandas as pd

# sort the rows of dataframe in descending
# order based on row index
rslt_df = df.sort_index(ascending = False)

# show the resultant Dataframe
rslt_df
```

**Sort a DataFrame based on column names**

```python
# import pandas library as pd
import pandas as pd

# sort a dataframe based on column names
rslt_df = df.sort_index(axis = 1)

# show the resultant Dataframe
rslt_df
```

```python
# sort a dataframe in descending
# order based on column names
rslt_df = df.sort_index(ascending = False, axis = 1)
 
# show the resultant Dataframe
rslt_df
```
